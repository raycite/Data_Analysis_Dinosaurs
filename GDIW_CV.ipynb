```python
# ======================================
# üß† Objective 1: Understand Fundamentals of Computer Vision and AI
# ======================================

from IPython.display import Markdown, display
import cv2
import matplotlib.pyplot as plt
import numpy as np

# --- Text introduction (shows nicely in Colab) ---
display(Markdown("""
# üß© Objective 1: Understanding the Fundamentals of Computer Vision and AI

### üëÅÔ∏è What is Computer Vision?
Computer Vision (CV) is a field of Artificial Intelligence that enables computers to **see, interpret, and understand visual information** ‚Äî much like humans do.

In this section, we‚Äôll explore how a computer ‚Äúsees‚Äù an image using Python.
"""))

# --- Step 1: Upload and display an image ---
display(Markdown("### üñºÔ∏è Step 1: Upload and View an Image"))

# üëâ Upload your image (for example, a photo or object)
from google.colab import files
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys())[0]

# Read and convert to RGB (OpenCV loads as BGR by default)
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image
plt.figure(figsize=(6, 4))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("üëÅÔ∏è How the Computer Sees Your Image")
plt.show()

# --- Step 2: Explore image as numerical data ---
display(Markdown("""
### üî¢ Step 2: Understanding the Image as Data
To a computer, an image is not just a photo ‚Äî it‚Äôs a **matrix (array) of numbers**.

Each pixel has **three values**: Red, Green, and Blue (RGB).
Let's take a look at the image‚Äôs structure.
"""))

# Display shape (height, width, color channels)
display(Markdown(f"**Image shape:** `{img_rgb.shape}` (Height, Width, Color Channels)"))

# Show pixel values from a small region
x, y = 50, 50  # pick a pixel
pixel_value = img_rgb[y, x]
display(Markdown(f"**Sample pixel value at (x={x}, y={y}):** `{pixel_value}`"))

# Show top-left corner zoom (50x50 region)
zoom = img_rgb[:50, :50]
plt.figure(figsize=(3, 3))
plt.imshow(zoom)
plt.axis('off')
plt.title("üîç Zoomed-in Section (Top-Left 50x50 pixels)")
plt.show()

display(Markdown("""
> üß† **Takeaway:**
> - Computers interpret images as **arrays of pixel intensities**, not as visual scenes.
> - Each pixel‚Äôs RGB values help the computer learn features such as edges, colors, and patterns ‚Äî the building blocks of computer vision models.
"""))
```

```python
# ======================================
# ‚ö™ Step 2: Convert to Grayscale
# ======================================

import numpy as np

display(Markdown("""
### Step 2: Convert the Image to Grayscale
Each pixel in a grayscale image contains **one intensity value** instead of three (R, G, B).
This makes it easier for algorithms to detect **edges and shapes**.
"""))

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,4))
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.title("Grayscale Image")
plt.show()

display(Markdown(f"**Image shape (grayscale):** `{gray.shape}`"))
```

```python
# ======================================
# üîé Compare Zoomed-Out vs Zoomed-In (Standalone)
# ======================================

from IPython.display import Markdown, display
import matplotlib.pyplot as plt
import cv2

display(Markdown("""
### Comparing Zoom Out vs Zoom In
When **zoomed out**, your eye blends pixels into smooth shapes.
When **zoomed in**, you see how every pixel intensity contributes to the image.
"""))

# Reload grayscale image
gray = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2GRAY)
h, w = gray.shape
zoom_region = gray[h//3:h//3+50, w//3:w//3+50]

fig, ax = plt.subplots(1, 2, figsize=(10,4))

# Zoomed-out (whole image)
ax[0].imshow(gray, cmap='gray')
ax[0].set_title("Zoom Out ‚Äì Smooth Image View")
ax[0].axis("off")

# Zoomed-in (small region)
ax[1].imshow(zoom_region, cmap='gray', interpolation='nearest')
ax[1].set_title("Zoom In ‚Äì Visible Pixels")
ax[1].axis("off")

plt.show()

display(Markdown("""
> üß† **Observation:**
> - Each square is a **pixel** (0 = black, 255 = white).
> - The computer always sees **numbers**, even when we see shapes.
> - Zooming out averages neighboring pixels into smooth forms.
"""))
```

```python
# ======================================
# üîé Compare Zoomed-Out vs Zoomed-In with Region Highlight
# ======================================

from IPython.display import Markdown, display
import matplotlib.pyplot as plt
import cv2

display(Markdown("""
### Comparing Zoom Out vs Zoom In
When **zoomed out**, the image appears smooth and continuous.
When **zoomed in**, individual **pixels** (tiny squares of intensity) become visible.
"""))

# Reload grayscale image
gray = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2GRAY)

# Define zoom region coordinates
h, w = gray.shape
x, y, size = w//3, h//3, 60
zoom_region = gray[y:y+size, x:x+size]

# Draw rectangle on full image to show zoom region
image_with_box = gray.copy()
cv2.rectangle(image_with_box, (x, y), (x+size, y+size), (255, 0, 0), 2)

# Plot side-by-side
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Zoomed-out (full image with red box)
ax[0].imshow(image_with_box, cmap='gray')
ax[0].set_title("Zoom Out ‚Äì Full Image (Region Highlighted)")
ax[0].axis("off")

# Zoomed-in (just that region, magnified)
ax[1].imshow(zoom_region, cmap='gray', interpolation='nearest')
ax[1].set_title("Zoom In ‚Äì Pixel Level View")
ax[1].axis("off")

plt.show()

display(Markdown("""
> üß† **Observation:**
> - The left image shows the full picture ‚Äî how humans perceive patterns.
> - The right image shows the raw data ‚Äî how computers perceive pixels.
> - Computer vision models are trained to recognize meaning from these pixel patterns.
"""))
```
